```python
import pandas as pd

df_raw = pd.read_csv('joined_master_sparse.csv')

# Remove entries that lack genres or descriptions
df = df_raw[df_raw['num_genres'] != 0]
df = df[df['num_descriptors'] != 0]

df.head()
```

```python
labeled_filenames = df['filename'].apply(lambda x: '../../EBBA_impressions/' + x).tolist()
labeled_filenames
```

```python
import shutil
import os


dest = '../labeled_images'

src_files = labeled_filenames
for image in src_files:
    shutil.copy(image, dest)
```
